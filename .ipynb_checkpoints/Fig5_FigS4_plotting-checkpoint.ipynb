# Fig 5, Fig S4 plotting

Plots the data associated with Figure 5 and Figure S4.

### Standard Imports

```python
# standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import matplotlib as mpl
from cycler import cycler
import pylab
from matplotlib.colors import LogNorm, Normalize

# Plotting parameters
sns.set_context("talk", font_scale=1, rc={"lines.linewidth": 2.0, 'lines.markersize': 5})
sns.set_style("ticks")
sns.set_style({"xtick.direction": "in","ytick.direction": "in"})
mpl.rc('axes', prop_cycle=(cycler('color', ['r', 'k', 'b','g','y','m','c']) ))
mpl.rc('text', usetex=False)
tw = 1.5
sns.set_style({"xtick.major.size": 12, "ytick.major.size": 12,
               "xtick.minor.size": 8, "ytick.minor.size": 8,
               'axes.labelsize': 48,
               'xtick.major.width': tw, 'xtick.minor.width': tw,
               'ytick.major.width': tw, 'ytick.minor.width': tw})
mpl.rc('xtick', labelsize=24) 
mpl.rc('ytick', labelsize=24)
mpl.rc('axes', linewidth=1.75)
plt.gcf().subplots_adjust(bottom=0.15)
sns.set_style({'axes.labelsize': 48, 'axes.titlesize': 48})
cm = 1/2.54


# Color variables
cb_green = (102/255, 194/255, 165/255, 0.5)
cb_orange = (252/255, 141/255, 98/255, 0.5)
cb_blue = (141/255, 160/255, 203/255, 0.5)
cb_darkblue = (141/255, 160/255, 203/255, 0.25)
alpha_grey = (215/255, 215/255, 215/255, 1.0)
alpha_val = 0.5
alpha_red = (alpha_val + 1/2, alpha_val, alpha_val)
alpha_black = (alpha_val, alpha_val, alpha_val)
alpha_blue = (alpha_val, alpha_val, alpha_val + 1/2)

%matplotlib inline
```

### Import data

```python
df = pd.read_excel('./data/CFU_counts.xlsx', sheet_name='Fig 5')
```

### Plot raw CFU data

This is not one of the figures in the manuscript.

```python
plt.figure(figsize=(10,18))
g = sns.FacetGrid(df, row='Sal (uM)',
                  sharex=False, sharey=True, margin_titles=True,
                  aspect=1, height=6 )

g.map(sns.barplot, 'Receiver gRNA', 'CFUs/mL', 'Strain',
      palette=[alpha_red, alpha_black, alpha_blue], errcolor='k', edgecolor='k',
      ci=None, estimator=stats.mstats.gmean,
      order=['A','B'],
      hue_order=['Sender', 'Receiver', 'Transconjugant'])
g.map(sns.swarmplot, 'Receiver gRNA', 'CFUs/mL', 'Strain',
      palette=['k'], edgecolor='k', s=10, dodge=True,
      order=['A','B'],
      hue_order=['Sender', 'Receiver', 'Transconjugant'],
     )

g.set(yscale='log',
      ylim=(1e4, 3e9)
     )
g.add_legend()

plt.show()
```

### Calculate transfer rates

```python
# Get transfer rates
rate_df = pd.DataFrame(columns=[
                                'Receiver gRNA',
                                'Sal (uM)',
                                'Transfer Rate',
                                'Penetrance',
                                ], dtype='float64')


reps = [1,2,3]
sites = ['A', 'B']
sals = [0, 100]

for s in range(len(sites)):
    for c in range(len(sals)):
        for r in range(len(reps)):
            rep = reps[r]
            site = sites[s]
            sal = sals[c]

            S = df[(df['Replicate'] == rep) &
                   (df['Receiver gRNA'] == site) &
                   (df['Sal (uM)'] == sal) &
                   (df['Strain'] == 'Sender')]['CFUs/mL'].values
            R = df[(df['Replicate'] == rep) &
                   (df['Receiver gRNA'] == site) &
                   (df['Sal (uM)'] == sal) &
                   (df['Strain'] == 'Receiver')]['CFUs/mL'].values
            T = df[(df['Replicate'] == rep) &
                   (df['Receiver gRNA'] == site) &
                   (df['Sal (uM)'] == sal) &
                   (df['Strain'] == 'Transconjugant')]['CFUs/mL'].values

            assert len(S) == 1
            assert len(R) == 1
            assert len(T) == 1

            S = S[0]
            R = R[0]
            T = T[0]

            rate = T/(S*R)
            pen= T/R
            
            tempdict = pd.DataFrame({'Replicate': rep,
                        'Receiver gRNA': site,
                        'Sal (uM)': sal,
                        'Transfer Rate': rate,
                        'Penetrance': pen},
                                    index=[0])

            rate_df = pd.concat([rate_df, tempdict], ignore_index=True)                           
```

### Plot transfer rates

```python
plt.figure(figsize=(20,8))
g = sns.FacetGrid(rate_df, row='Sal (uM)',
                  sharex=False, sharey=True, margin_titles=True,
                  aspect=1.0, height=5 )

g.map(sns.barplot, 'Receiver gRNA', 'Transfer Rate',
      palette=[cb_blue], ci=None, edgecolor='k',
      order=['A','B'], estimator=stats.mstats.gmean,
     )
g.map(sns.swarmplot, 'Receiver gRNA', 'Transfer Rate',
      palette=['k'], edgecolor='k', s=10,
      order=['A','B']
     )

g.set(yscale='log',
      ylim=(1e-13, 3e-9)
     )
g.add_legend()

plt.savefig('./figs/Fig_5C.pdf', bbox_inches='tight')

plt.show()
```

### Calculate fold changes in transfer rates

```python
AA1 = rate_df[(rate_df['Receiver gRNA'] == 'A') &
                (rate_df['Sal (uM)'] == 100)]['Transfer Rate'].values
AB1 = rate_df[(rate_df['Receiver gRNA'] == 'B') &
                (rate_df['Sal (uM)'] == 100)]['Transfer Rate'].values

AA0 = rate_df[(rate_df['Receiver gRNA'] == 'A') &
                (rate_df['Sal (uM)'] == 0)]['Transfer Rate'].values
AB0 = rate_df[(rate_df['Receiver gRNA'] == 'B') &
                (rate_df['Sal (uM)'] == 0)]['Transfer Rate'].values

gmAA1 = stats.mstats.gmean(AA1)
gmAB1 = stats.mstats.gmean(AB1)
gmAA0 = stats.mstats.gmean(AA0)
gmAB0 = stats.mstats.gmean(AB0)

print(f'The fold change in transfer rate without editing was {gmAB0/gmAA0}.')
print(f'The fold change in transfer rate with editing was {gmAA1/gmAB1}.')
```

### Peform statistical analysis on transfer rate differences

We will now perform a paired t-test to ask whether the geometric mean of the transfer rate associated with a given message plasmid condition (unedited or edited) differs when paired with an on-target or off-target receiver.

```python
t0, p0 = stats.ttest_rel(np.log10(AA0), np.log10(AB0))
t1, p1 = stats.ttest_rel(np.log10(AA1), np.log10(AB1))

print(f'The p value of the test on the unedited senders is {p0:.3f}.')
print(f'The p value of the test on the edited senders is {p1:.3f}.')
```

## Import Fig 3 data for comparison (for Fig S4)

```python
df_block = pd.read_excel('./data/CFU_counts.xlsx', sheet_name='Fig 3')
```

### Calculate transfer rates of Fig 3 data

```python
# Get transfer rates
rate_df_block = pd.DataFrame(columns=[
                                'Message Address',
                                'Receiver gRNA',
                                'Cin (uM)',
                                'Transfer Rate',
                                ], dtype='float64')

reps = [1,2,3]
sites = ['A', 'B']
cins = [0, 1]

for a in range(len(sites)):
    for s in range(len(sites)):
        for c in range(len(cins)):
            for r in range(len(reps)):
                address = sites[a]
                rep = reps[r]
                site = sites[s]
                cin = cins[c]

                S = df_block[(df_block['Replicate'] == rep) &
                       (df_block['Message Address'] == address) &
                       (df_block['Receiver gRNA'] == site) &
                       (df_block['Cin (uM)'] == cin) &
                       (df_block['Strain'] == 'Sender')]['CFUs/mL'].values
                R = df_block[(df_block['Replicate'] == rep) &
                       (df_block['Message Address'] == address) &
                       (df_block['Receiver gRNA'] == site) &
                       (df_block['Cin (uM)'] == cin) &
                       (df_block['Strain'] == 'Receiver')]['CFUs/mL'].values
                T = df_block[(df_block['Replicate'] == rep) &
                       (df_block['Message Address'] == address) &
                       (df_block['Receiver gRNA'] == site) &
                       (df_block['Cin (uM)'] == cin) &
                       (df_block['Strain'] == 'Transconjugant')]['CFUs/mL'].values

                assert len(S) == 1
                assert len(R) == 1
                assert len(T) == 1

                S = S[0]
                R = R[0]
                T = T[0]

                rate = T/(S*R)

                tempdict = pd.DataFrame({'Replicate': rep,
                            'Message Address': address,
                            'Receiver gRNA': site,
                            'Cin (uM)': cin,
                            'Transfer Rate': rate}, index=[0])

                rate_df_block = pd.concat([rate_df_block, tempdict], ignore_index=True)
```

## Calculate the geometric means and standard deviations of Fig 3, 5 data

```python
fig3_summary_df = pd.DataFrame(columns=[
    'Message Address',
    'Receiver gRNA',
    'Geomean Transfer Rate',
    'GeoSTD Transfer Rate',
])
fig5_summary_df = pd.DataFrame(columns=[
    'Message Address',
    'Receiver gRNA',
    'Geomean Transfer Rate',
    'GeoSTD Transfer Rate',
])

gRNAs = ['A', 'B']
messages = ['A', 'B']
sals = [0, 100]

for g in range(len(gRNAs)):
    for m in range(len(messages)):
        gRNA = gRNAs[g]
        message = messages[m]
        sal = sals[m]
        
        fig3_rates = rate_df_block[(rate_df_block['Message Address'] == message) &
                                   (rate_df_block['Receiver gRNA'] == gRNA) &
                                   (rate_df_block['Cin (uM)'] == 1)]['Transfer Rate'].values
        fig5_rates = rate_df[(rate_df['Receiver gRNA'] == gRNA) &
                             (rate_df['Sal (uM)'] == sal)]['Transfer Rate'].values
        
        assert len(fig3_rates) == 3
        assert len(fig5_rates) == 3
        
        fig3_gmean = stats.mstats.gmean(fig3_rates)
        fig3_gstd = stats.gstd(fig3_rates)
        fig5_gmean = stats.mstats.gmean(fig5_rates)
        fig5_gstd = stats.gstd(fig5_rates)
        
        tempdf_3 = pd.DataFrame({
                               'Message Address': message,
                               'Receiver gRNA': gRNA,
                               'Geomean Transfer Rate': fig3_gmean,
                               'GeoSTD Transfer Rate': fig3_gstd}, index=[0])
        tempdf_5 = pd.DataFrame({
                               'Message Address': message,
                               'Receiver gRNA': gRNA,
                               'Geomean Transfer Rate': fig5_gmean,
                               'GeoSTD Transfer Rate': fig5_gstd}, index=[0])
        fig3_summary_df = pd.concat([fig3_summary_df, tempdf_3],
                                  ignore_index=True)
        fig5_summary_df = pd.concat([fig5_summary_df, tempdf_5],
                                  ignore_index=True)
        
print(fig3_summary_df)
print(fig5_summary_df)
```

Since both dataframes are now in the same ordering, we can directly extract the values as they are.

```python
fig3_rates = fig3_summary_df['Geomean Transfer Rate'].values
fig5_rates = fig5_summary_df['Geomean Transfer Rate'].values

fig3_stds = fig3_summary_df['GeoSTD Transfer Rate'].values
fig5_stds = fig5_summary_df['GeoSTD Transfer Rate'].values

fig3_errorlocs_lower = fig3_rates / fig3_stds
fig3_errorlocs_upper = fig3_rates * fig3_stds
fig5_errorlocs_lower = fig5_rates / fig5_stds
fig5_errorlocs_upper = fig5_rates * fig5_stds

plt.figure(figsize=(7,6))

# Plot geometric means
plt.scatter(fig3_rates[[0,2,3]],
            fig5_rates[[0,2,3]],
            s=50,
            c='k',
           )
plt.scatter(fig3_rates[1],
            fig5_rates[1],
            s=50,
            c='r',
           )

# Plot geometric STDs
plt.vlines(fig3_rates[[0,2,3]],
           fig5_errorlocs_lower[[0,2,3]],
           fig5_errorlocs_upper[[0,2,3]],
           color='k',
           linewidth=1,)
plt.hlines(fig5_rates[[0,2,3]],
           fig3_errorlocs_lower[[0,2,3]],
           fig3_errorlocs_upper[[0,2,3]],
           color='k',
           linewidth=1,)
plt.vlines(fig3_rates[1],
           fig5_errorlocs_lower[1],
           fig5_errorlocs_upper[1],
           color='r',
           linewidth=1,)
plt.hlines(fig5_rates[1],
           fig3_errorlocs_lower[1],
           fig3_errorlocs_upper[1],
           color='r',
           linewidth=1,)
plt.gca().set_xscale('log')
plt.gca().set_yscale('log')
plt.xlabel('Transfer Rate (Fig 3)')
plt.ylabel('Transfer Rate (Fig 5)')

# find line of best fit through 3 'colinear' points
a, b = np.polyfit(np.log10(fig3_rates[[0,2,3]].astype(float)),
                  np.log10(fig5_rates[[0,2,3]].astype(float)),
                  1
                 )

xmin, xmax = plt.gca().get_xlim()
ymin, ymax = plt.gca().get_ylim()

# Plot line of best fit
plt.plot([1e-12, 1e-8],
         [10**(a*(-12) + b), 10**(a*(-8) + b)],
         'k--')
# Plot logy=logx
plt.plot([1e-13, 1e-8],
         [1e-13, 1e-8],
         'k-')

# Label Transfers
plt.text(fig3_rates[0]*1.15,
         fig5_rates[0]*.5,
         'A -> A',
         fontsize=14,)
plt.text(fig3_rates[2]*.65,
         fig5_rates[2]*.35,
         'A -> B',
         fontsize=14,)
plt.text(fig3_rates[3]*1.15,
         fig5_rates[3]*.55,
         'B -> B',
         fontsize=14,)
plt.text(fig3_rates[1]*.65,
         fig5_rates[1]*.4,
         'B -> A',
         fontsize=14,
         c='r')
plt.text(3.2e-11,
         9.2e-12,
         f'log(y) = {a:.3f} log(x) - {b:.3f}',
         rotation = 45*a,
         fontsize=14,
        )

plt.gca().set_xlim((6e-13, 8e-9))
plt.gca().set_ylim((6e-13, 8e-9))

plt.tight_layout()
plt.savefig('./figs/Fig_S4b.pdf')
plt.show()
```

### Calculate fold difference in observed Fig 5 rate from expected Fig 5 rate

For the B-to-A transfer marked in red above, we can calculate the fold change between the observed Fig 5 transfer rate and the expected Fig 5 transfer rate from the line of best fit.

```python
predicted_SU_fig4 = 10**(a*np.log10(fig3_rates[1]) + b)
fc = predicted_SU_fig4 / fig5_rates[1]

print(f'The fold change between the observed and expected Fig 5 transfer rate for the B->A transfer is {fc}.')
```

### Calculate the global decrease in transfer rate in Fig 5 compared to Fig 3

```python
AA_drop = fig3_rates[0]/fig5_rates[0]
AB_drop = fig3_rates[2]/fig5_rates[2]
BB_drop = fig3_rates[3]/fig5_rates[3]

mean_global_drop = stats.mstats.gmean([AA_drop, AB_drop, BB_drop])

print(f'The Fig 5 B->B transfer is {BB_drop:.2f}-fold lower than in Fig 3.')
print(f'The Fig 5 A->A transfer is {AA_drop:.2f}-fold lower than in Fig 3.')
print(f'The Fig 5 A->B transfer is {AB_drop:.2f}-fold lower than in Fig 3.')
print(f'The mean of the above fold changes is {np.mean([AA_drop, AB_drop, BB_drop]):.2f}-fold.')
```
