### Calculate Transfer Rates

Calculate $T/(S*R)$ from the above data for each condition, where $T$ is the transconjugant density, $S$ is the sender density, and $R$ is the total receiver density. Also calculate the geometric mean transfer rate of the three replicates for each condition.

```python
# Get transfer rates
rate_df = pd.DataFrame(columns=[
                                'Message Address',
                                'Receiver gRNA',
                                'Cin (uM)',
                                'Transfer Rate',
                                ], dtype='float64')

reps = [1,2,3]
sites = ['A', 'B']
cins = [0, 1]

for a in range(len(sites)):
    for s in range(len(sites)):
        for c in range(len(cins)):
            for r in range(len(reps)):
                address = sites[a]
                rep = reps[r]
                site = sites[s]
                cin = cins[c]

                S = df[(df['Replicate'] == rep) &
                       (df['Message Address'] == address) &
                       (df['Receiver gRNA'] == site) &
                       (df['Cin (uM)'] == cin) &
                       (df['Strain'] == 'Sender')]['CFUs/mL'].values
                R = df[(df['Replicate'] == rep) &
                       (df['Message Address'] == address) &
                       (df['Receiver gRNA'] == site) &
                       (df['Cin (uM)'] == cin) &
                       (df['Strain'] == 'Receiver')]['CFUs/mL'].values
                T = df[(df['Replicate'] == rep) &
                       (df['Message Address'] == address) &
                       (df['Receiver gRNA'] == site) &
                       (df['Cin (uM)'] == cin) &
                       (df['Strain'] == 'Transconjugant')]['CFUs/mL'].values

                assert len(S) == 1
                assert len(R) == 1
                assert len(T) == 1

                S = S[0]
                R = R[0]
                T = T[0]

                rate = T/(S*R)

                tempdict = pd.DataFrame({'Replicate': rep,
                            'Message Address': address,
                            'Receiver gRNA': site,
                            'Cin (uM)': cin,
                            'Transfer Rate': rate}, index=[0])

                rate_df = pd.concat([rate_df, tempdict], ignore_index=True)

                    
```

```python
# Plot transfer rates

plt.figure(figsize=(4*cm,4*cm))
g = sns.FacetGrid(rate_df, row='Message Address',
                  sharex=False, sharey=True, margin_titles=True,
                  aspect=.8, 
                  height=6,
                 )

g.map(sns.barplot, 'Receiver gRNA', 'Transfer Rate', 'Cin (uM)',
      palette=[alpha_grey, cb_darkblue], alpha=None, ci=None, edgecolor='k',
      order=['A','B'], hue_order=[0,1], estimator=stats.mstats.gmean,
     )
g.map(sns.swarmplot, 'Receiver gRNA', 'Transfer Rate', 'Cin (uM)',
      palette=['k'], edgecolor='k', s=10, dodge=True,
      order=['A','B'], hue_order=[0,1],
     )

g.set(yscale='log',
      ylim=(1e-12, 1e-8)
     )

g.add_legend(title='Cin (uM)')

plt.savefig('./figs/Fig_3C.pdf', bbox_inches='tight')

plt.show()
```

### Calculate fold change in transfer rates across different conditions

```python
# Extract geometric mean transfer rate for each condition 
AA1 = rate_df[(rate_df['Message Address'] == 'A') &
                (rate_df['Receiver gRNA'] == 'A') &
                (rate_df['Cin (uM)'] == 1)
               ]['Transfer Rate'].values
AB1 = rate_df[(rate_df['Message Address'] == 'A') &
                (rate_df['Receiver gRNA'] == 'B') &
                (rate_df['Cin (uM)'] == 1)
               ]['Transfer Rate'].values
BA1 = rate_df[(rate_df['Message Address'] == 'B') &
                (rate_df['Receiver gRNA'] == 'A') &
                (rate_df['Cin (uM)'] == 1)
               ]['Transfer Rate'].values
BB1 = rate_df[(rate_df['Message Address'] == 'B') &
                (rate_df['Receiver gRNA'] == 'B') &
                (rate_df['Cin (uM)'] == 1)
               ]['Transfer Rate'].values

AA0 = rate_df[(rate_df['Message Address'] == 'A') &
                (rate_df['Receiver gRNA'] == 'A') &
                (rate_df['Cin (uM)'] == 0)
               ]['Transfer Rate'].values
AB0 = rate_df[(rate_df['Message Address'] == 'A') &
                (rate_df['Receiver gRNA'] == 'B') &
                (rate_df['Cin (uM)'] == 0)
               ]['Transfer Rate'].values
BA0 = rate_df[(rate_df['Message Address'] == 'B') &
                (rate_df['Receiver gRNA'] == 'A') &
                (rate_df['Cin (uM)'] == 0)
               ]['Transfer Rate'].values
BB0 = rate_df[(rate_df['Message Address'] == 'B') &
                (rate_df['Receiver gRNA'] == 'B') &
                (rate_df['Cin (uM)'] == 0)
               ]['Transfer Rate'].values

gmAA1 = stats.mstats.gmean(AA1)
gmAB1 = stats.mstats.gmean(AB1)
gmBA1 = stats.mstats.gmean(BA1)
gmBB1 = stats.mstats.gmean(BB1)
gmAA0 = stats.mstats.gmean(AA0)
gmAB0 = stats.mstats.gmean(AB0)
gmBA0 = stats.mstats.gmean(BA0)
gmBB0 = stats.mstats.gmean(BB0)

print('When Cas9 blocking was activated...')
print(f'The A message was biased towards its programmed recipient by {gmAB1/gmAA1}-fold.')
print(f'The B message was biased towards its programmed recipient by {gmBA1/gmBB1}-fold.')
print('\nWhen Cas9 blocking was not activated...')
print(f'The fold change in transfer rates between recipients for the A message was {gmAB0/gmAA0}-fold.')
print(f'The fold change in transfer rates between recipients for the B message was {gmBB0/gmBA0}-fold.')
```

### Perform statistics on transfer rates

We will perform a paired t-test to assess whether the geometric mean of the distribution of transfer rates, for a given sender-receiver pair, changed between the blocking-uninduced and the blocking-induced conditions.

We will assume the data are log-normally distributed, so we will satisfy the normality assumption for the t test by log-transforming the data. There are 2 degrees of freedom because $n=3$.

```python
tAA, pAA = stats.ttest_rel(np.log10(AA0), np.log10(AA1))
tAB, pAB = stats.ttest_rel(np.log10(AB0), np.log10(AB1))
tBA, pBA = stats.ttest_rel(np.log10(BA0), np.log10(BA1))
tBB, pBB = stats.ttest_rel(np.log10(BB0), np.log10(BB1))

print(f'The p value of the test on the A->A transfer is {pAA:.3f}.')
print(f'The p value of the test on the A->B transfer is {pAB:.3f}.')
print(f'The p value of the test on the B->A transfer is {pBA:.3f}.')
print(f'The p value of the test on the B->B transfer is {pBB:.3f}.')
```

### Calculate fractional receiver conversion values for Fig S2

Receiver penetrance, or fractional receiver conversion, is defined as $T/R$ and in principle takes a value between 0 and 1. Although in practice the fact that $T$ and $R$ are obtained from independent experimental measurements means that this ratio may exceed 1.

```python
# Get penetrance values

penetrance_df = pd.DataFrame(columns=[
                                'Message Address',
                                'Receiver gRNA',
                                'Cin (uM)',
                                'Penetrance',
                                ], dtype='float64')

reps = [1,2,3]
sites = ['A', 'B']
cins = [0, 1]

for a in range(len(sites)):
    for s in range(len(sites)):
        for c in range(len(cins)):
            for r in range(len(reps)):
                address = sites[a]
                rep = reps[r]
                site = sites[s]
                cin = cins[c]

                S = df[(df['Replicate'] == rep) &
                       (df['Message Address'] == address) &
                       (df['Receiver gRNA'] == site) &
                       (df['Cin (uM)'] == cin) &
                       (df['Strain'] == 'Sender')]['CFUs/mL'].values
                R = df[(df['Replicate'] == rep) &
                       (df['Message Address'] == address) &
                       (df['Receiver gRNA'] == site) &
                       (df['Cin (uM)'] == cin) &
                       (df['Strain'] == 'Receiver')]['CFUs/mL'].values
                T = df[(df['Replicate'] == rep) &
                       (df['Message Address'] == address) &
                       (df['Receiver gRNA'] == site) &
                       (df['Cin (uM)'] == cin) &
                       (df['Strain'] == 'Transconjugant')]['CFUs/mL'].values

                assert len(S) == 1
                assert len(R) == 1
                assert len(T) == 1

                S = S[0]
                R = R[0]
                T = T[0]

                pen = T/R
                
                tempdict_pen = pd.DataFrame({'Replicate': rep,
                            'Message Address': address,
                            'Receiver gRNA': site,
                            'Cin (uM)': cin,
                            'Penetrance': pen},
                                            index=[0])

                penetrance_df = pd.concat([penetrance_df, tempdict_pen],
                                          ignore_index=True)


```

### Plot receiver penetrance for no-block condition

```python
# Plot transfer rates

plt.figure(figsize=(4*cm,4*cm))
g = sns.FacetGrid(penetrance_df, row='Message Address',
                  sharex=False, sharey=True, margin_titles=True,
                  aspect=.8, 
                  height=6,
                 )

g.map(sns.barplot, 'Receiver gRNA', 'Penetrance', 'Cin (uM)',
      palette=[alpha_grey, cb_darkblue], alpha=None, ci=None, edgecolor='k',
      order=['A','B'], hue_order=[0,1], estimator=stats.mstats.gmean,
     )
g.map(sns.swarmplot, 'Receiver gRNA', 'Penetrance', 'Cin (uM)',
      palette=['k'], edgecolor='k', s=10, dodge=True,
      order=['A','B'], hue_order=[0,1],
     )

g.set(ylabel='Fraction Receivers Converted')
g.set(yscale='log',
     ylim=(9e-4,3e0))

g.add_legend(title='Cin (uM)')

plt.savefig('./figs/Fig_S2.pdf', bbox_inches='tight')

plt.show()
```

### Calculate average receiver conversion values

To calculate the mean receiver conversion of all the on-target conditions, we can use the fact that all on-target conversions were above 0.1 and all off-target conversions were below 0.1.

```python
pens = penetrance_df['Penetrance'].values
ontarget_pens = pens[pens > 0.1]

assert len(ontarget_pens) == 18

print(f'The geometric mean fractional receiver conversion across all on-target conditions was {stats.mstats.gmean(ontarget_pens)}.')
```

To calculate the mean receiver conversion from the conditions where the blocking system was not induced, we can slice those conditions out directly from the dataframe.

```python
mean_penetrance = stats.mstats.gmean(penetrance_df[penetrance_df['Cin (uM)'] == 0]['Penetrance'].values)

print(f'The geometric mean fractional receiver conversion across the four uninduced conditions was {mean_penetrance}.')
```
