### Import data

Note that the '50 uL' initial condition is allowed to grow undisturbed for 6 hours but after this point is diluted and allowed to regrow for another 6 hours. This second re-growth period is not included in Fig S5 as it is not relevant to the conclusion of the figure.

```python
df = pd.read_excel('./data/CFU_counts.xlsx', sheet_name='Fig S5')
```

```python
df_high = df[df['Initial Concentration'] == 'Stationary']
df_low = df[df['Initial Concentration'] == '50 nL']
df_med = df[df['Initial Concentration'] == '50 uL']

fig, ax = plt.subplots(1,3,figsize=(16,4), sharey=True)

ax[0].plot(df_low['Time (hr)'].values,
           df_low['Senders'].values,
           label='Senders',
           c='r')
ax[0].plot(df_low['Time (hr)'].values,
           df_low['Receivers'].values,
           label='Receivers',
           c='k')
ax[0].plot(df_low['Time (hr)'].values,
           df_low['Transconjugants'].values,
           label='Transconjugants',
           c='b')
ax[1].plot(df_med['Time (hr)'].values[:7],
           df_med['Senders'].values[:7],
           label='Senders',
           c='r')
ax[1].plot(df_med['Time (hr)'].values[:7],
           df_med['Receivers'].values[:7],
           label='Receivers',
           c='k')
ax[1].plot(df_med['Time (hr)'].values[:7],
           df_med['Transconjugants'].values[:7],
           label='Transconjugants',
           c='b')
ax[2].plot(df_high['Time (hr)'].values,
           df_high['Senders'].values,
           label='Senders',
           c='r')
ax[2].plot(df_high['Time (hr)'].values,
           df_high['Receivers'].values,
           label='Receivers',
           c='k')
ax[2].plot(df_high['Time (hr)'].values,
           df_high['Transconjugants'].values,
           label='Transconjugants',
           c='b')

ax[0].scatter(df_low['Time (hr)'].values,
           df_low['Senders'].values,
           c='r')
ax[0].scatter(df_low['Time (hr)'].values,
           df_low['Receivers'].values,
           c='k')
ax[0].scatter(df_low['Time (hr)'].values,
           df_low['Transconjugants'].values,
           c='b')
ax[1].scatter(df_med['Time (hr)'].values[:7],
           df_med['Senders'].values[:7],
           c='r')
ax[1].scatter(df_med['Time (hr)'].values[:7],
           df_med['Receivers'].values[:7],
           c='k')
ax[1].scatter(df_med['Time (hr)'].values[:7],
           df_med['Transconjugants'].values[:7],
           c='b')
ax[2].scatter(df_high['Time (hr)'].values,
           df_high['Senders'].values,
           c='r')
ax[2].scatter(df_high['Time (hr)'].values,
           df_high['Receivers'].values,
           c='k')
ax[2].scatter(df_high['Time (hr)'].values,
           df_high['Transconjugants'].values,
           c='b')

ax[0].set_yscale('log')
ax[1].set_yscale('log')
ax[2].set_yscale('log')

ax[2].legend(loc='lower right')

ax[0].set_ylim((1e0, 3e10))

for i in range(3):
    ax[i].set_xlabel('Time (hr)')
ax[0].set_ylabel('CFUs/mL')

plt.tight_layout()
plt.savefig('./figs/Fig_S6.pdf')

plt.show()
```
